**Task I am doing:** I am building a linear regression model using the `canada_per_capita_income.csv` dataset to predict Canada's per-capita income for the year 2020.

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/canada_per_capita_income.csv')
df=df.rename(columns={"per capita income (US$)":"per_capita_income"})
df
```

```python
plt.xlabel('year')
plt.ylabel('per_capita_income (US$) ')
plt.scatter(df.year,df.per_capita_income,color='red',marker='+')
plt.show()
```

```python
new_df = df.drop('per_capita_income',axis='columns')
new_df
```

```python
per_capita_income=df.per_capita_income
per_capita_income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
temp = 2020
new_year = pd.DataFrame([[temp]], columns=['year'])
predicted_income = reg.predict(new_year)
print(f"Predicted Income for year {temp} : $ {predicted_income[0]}")
```

```python
reg.coef_
```

```python
# Plot the actual data points
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df.year, df.per_capita_income, color='red', marker='+',label="Actual Data")

# Plot the predicted point 
plt.scatter(temp,predicted_income,color='black',marker='.',label='Predicted Point')

# Generate regression line
plt.plot(df.year, reg.predict(new_df), color='blue',label="Regression Line")

plt.legend()
plt.show()
```
